```python
import opendatasets as od
import pandas as pd
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
od.download("https://www.kaggle.com/datasets/mjshri23/life-expectancy-and-socio-economic-world-bank")
```

```python
df_raw = pd.read_csv("life-expectancy-and-socio-economic-world-bank/life expectancy.csv",encoding="ISO=8859-1")
df_raw.info()
```

```python
df_raw.head()
```

```python
df_raw.isna().sum()
```

```python
dfnn = df_raw.dropna(axis=0)
```

```python
dfnn.info()
```

```python
dfnn.shape
```

```python
dfnn.rename(columns={"Country Name":"Country"},inplace=True)
dfnn.info()
```

```python
dfnn.set_index('Country', inplace=True)
```

```python
dfnn.corr()
```

```python
dfnn.head()
```

```python
import numpy as np
count,bin_edges = np.histogram(dfnn[['CO2','Corruption']], 20)
xmin = bin_edges[0] - 10
xmax = bin_edges[-1] + 10

dfnn[['CO2','Corruption']].plot(kind='hist',
                                  figsize=(10,6),
                                  alpha=0.25,
                                  stacked=False,
                                  bins=20,
                                  xticks=bin_edges,
                                  xlim=(xmin,xmax)
                                 )
plt.xlabel('No. of CO2 and Communicable')
plt.ylabel('No. of Countries')
plt.title('A Histogram')
plt.show()
```
